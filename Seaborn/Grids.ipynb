```python
import seaborn as sns
%matplotlib inline

iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)#gives empty grids.
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.displot)
#g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
